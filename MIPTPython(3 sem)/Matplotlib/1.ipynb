```python
def read_file(file):
    data = []
    with open(file) as file:
        data = file.read()
    return data
```

```python
def make_points(data_set, cnt):
    x, y = [], []
    for i in range(1, 2*int(cnt) + 1, 2):
        x.append(data_set[i])
        y.append(data_set[i + 1])
    return [x, y]
```

```python
files = ["001.dat", "002.dat", "003.dat", "004.dat", "005.dat"]
data_sets = [[float(el) for el in read_file(file).split()] for file in files]
counters = [set_[0] for set_ in data_sets]
counters
```

```python
def read(file):
    data = []
    with open(file, "r") as f:
        data = f.readlines()
    return data
```

```python
FILE = "second.txt"
data = read(FILE)
# for line in data:
#     print(line.split())
data = [line.split() for line in data]
data = [[float(el) for el in current] for current in data]
len(data)
```

```python
fig, ax = plt.subplots(3, 2)
fig.set_figheight(15)
fig.set_figwidth(15)
ax[0][0].plot(data[0], data[1], color="green")
ax[0][1].plot(data[2], data[3], color="black")  
ax[1][0].plot(data[4], data[5], color="orange")
ax[1][1].plot(data[6], data[7], color="red")
ax[2][0].plot(data[8], data[9], color="yellow")  
ax[2][1].plot(data[10], data[11], color="blue")
for i in range(3):
    for j in range(2):
        ax[i][j].grid()
        ax[i][j].axis([0, 20, -10, 15])
        ax[i][j].set_title("Frame {}".format(str(2 * i + j)))
```

```python
import csv
from collections import Counter
```

```python
def read_csv(file):
    data = []
    with open(file, "r") as f:
        csv_reader = csv.reader(f, delimiter=';')
        for row in csv_reader:
            data.append(row)
    return data
```

```python
file3 = "students.csv"
to_work = read_csv(file3)
```

```python
first_bar = {}
for elem in to_work:
    if elem[0] not in first_bar:
        first_bar[elem[0]] = []
    first_bar[elem[0]].append(int(elem[2]))
```

```python
first_bar
datset = Counter(first_bar["prep1"])
```

```python
labels, sizes = [datset[key] for key in datset], [key for key in datset]
explode = (0.1, 0, 0, 0)  # explode 1st slice
# Plot
plt.pie(sizes, labels=labels)
plt.axis('equal')
plt.show()
```
